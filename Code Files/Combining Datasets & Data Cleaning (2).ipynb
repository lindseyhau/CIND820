```python
toronto_w = [w1, w2, w3, w4, w5, w6, w7, w8, w9, w10, w11, w12]
weather = pd.concat(toronto_w).reset_index(drop=True)
weather.head()
```

```python
# Number of observations and features
weather.shape
```

```python
# Datatypes
weather.info()
```

```python
# Classify wind as strong wind using boolean column. 1 if >= 60 km/h, 0 if < 60 km/h
weather['Strong Wind'] = (weather['Wind Spd (km/h)']>=60).astype(int)
```

```python
# Drop features not needed
weather.drop(['Longitude (x)', 
                 'Latitude (y)', 
                 'Station Name', 
                 'Climate ID',
                 'Year', 
                 'Month', 
                 'Day', 
                 'Time',
                 'Temp Flag', 
                 'Dew Point Temp Flag', 
                 'Rel Hum Flag', 
                 'Wind Dir (10s deg)',
                 'Wind Dir Flag',
                 'Wind Spd (km/h)',
                 'Wind Spd Flag', 
                 'Visibility Flag',
                 'Stn Press Flag', 
                 'Hmdx Flag',
                 'Wind Chill Flag'], axis=1, inplace=True)
```

```python
weather.head()
```

```python
weather.Weather.unique()
```

```python
# Try to simplify weather feature. Transform weather into a sparse matrix
from sklearn.feature_extraction.text import CountVectorizer
```

```python
weather['Weather'].fillna(value='clear', inplace=True) # fill in null values with stand in value
```

```python
weather_code = CountVectorizer(token_pattern=r'(?u)[a-zA-Z][a-z ]+')
weather_code_f_t = weather_code.fit_transform(weather['Weather'])
print(weather_code.get_feature_names())
```

```python
# Instantiate into dataframe and drop stand in value
weather_code_df = pd.DataFrame(columns=weather_code.get_feature_names(), 
                               data=weather_code_f_t.toarray())
weather_code_df.drop(['clear'], axis=1, inplace=True)
weather_code_df.columns = map(str.title, weather_code_df.columns)
```

```python
# Overwrite pervious weather with weather_code and drop weather feature
weather = pd.concat([weather, weather_code_df], join='inner', axis=1)
weather.drop(['Weather'], axis=1, inplace=True)
```

```python
weather.head()
```

```python
weather.info()
```

```python
# Check for duplicates
weather.duplicated().sum()
```

```python
# Check for null values
weather.isna().sum()
```

```python
# Handle missing values. Impute hmdx and wind chill with 0. Forwardfill any remaining null values
weather.fillna({'Hmdx': 0, 'Wind Chill': 0}, inplace=True)
weather.fillna(method='ffill', inplace=True)
```

```python
weather.isna().sum()
```

```python
# Convert Date/Time to DateTime format
weather['Date/Time'] = pd.to_datetime(weather['Date/Time'], yearfirst=True, infer_datetime_format=True)
```

```python
weather.info()
```

## Combine Bike Share & Weather Data

```python
bike_share2 = bike_share.copy()
```

```python
# Format date similar to weather data
bike_share2['trip_start_time'] = bike_share2['trip_start_time'].dt.strftime('%Y-%m-%d %H:00')
```

```python
bike_share2.head()
```

```python
# Group trips by date and hour
dates = {}
for d in bike_share2.trip_start_time:
    if d not in dates:
        dates[d] = 1
    else:
        dates[d] += 1
```

```python
df = pd.DataFrame.from_dict(dates, orient = "index")
df['date'] = df.index
df['trips'] = df.iloc[:,0]
```

```python
df.head()
```

```python
df2 = pd.DataFrame(df.date)
df2['trips'] = df['trips']
df2.reset_index(drop = True, inplace = True)
```

```python
df2.head()
```

```python
df2['date'] = df2['date'].astype('datetime64')
```

```python
# Join weather data
joined = df2.join(weather.set_index('Date/Time'), on='date')
```

```python
joined.head()
```

```python
# Re-add weekend and holiday boolean columns
joined['Weekend'] = ((joined['date'].dt.dayofweek) > 5).astype(int)
joined['Holiday'] = pd.to_datetime(joined['date']).dt.date.isin(ca_holidays).astype(int)
```

```python
# Extract hour and month from date 
joined['Hour'] = joined['date'].apply(lambda time: time.hour)
joined['Month'] = joined['date'].apply(lambda time: time.month)
```
